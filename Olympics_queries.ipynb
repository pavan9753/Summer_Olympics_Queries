# Summer Olympics Data Analysis Assignment

```python
# importing required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Readind the csv file with inbuilt function(read_csv) with pandas

```python
df=pd.read_csv("summer.csv")
df.head()
```

# 1.Total number of cities in which Summer Olympics held so far

```python
x=0
# Using basic for loop
for i in df['City'].unique():
  x+=1 
print("The number of cities in which olympics were held : ",x)
# Using set
print("The number of cities in which olympics were held : ",len(set(df['City'])))
# using Unique function
print("The number of cities in which olympics were held : ",len(df['City'].unique()))
```

# 2.Which sport is having most number of Gold Medals so far? (Top 5)

```python
data=[]
for sport in df['Sport'].unique():
  te=df[df['Sport']==sport]
  data.append([sport,len(te[te['Medal']=='Gold'])])
pd.DataFrame(data,columns=['Sport','No of gold medals']).sort_values(by='No of gold medals',ascending=False).head().plot(x='Sport',y='No of gold medals',kind='bar',title='Sport with most number of Gold medals',color='green',figsize=(10,5))


  
```

#3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]
for sport in df['Sport'].unique():
  data.append([sport,len(df[df['Sport']==sport])])
pd.DataFrame(data,columns=['Sport','No of medals']).sort_values(by='No of medals',ascending=False).head().plot(x='Sport',y='No of medals',title='Sport with most number of medals',kind='bar',figsize=(8,5))
```

# 4. Which player has won most number of medals? (Top 5)

```python
data=[]
for Athlete in df['Athlete'].unique():
  data.append([Athlete,len(df[df['Athlete']==Athlete])])
pd.DataFrame(data,columns=['Athlete','No of medals']).sort_values(by='No of medals',ascending=False).head().plot(x='Athlete',y='No of medals',kind='bar',title='Player with number of medals',figsize=(9,5))

  
```

# 5. Which player has won most number of Gold Medals ? (Top 5)

```python
data=[]
for Athlete in df['Athlete'].unique():
  te=df[df['Athlete']==Athlete]
  data.append([Athlete,len(te[te['Medal']=='Gold'])])
#print(data[0:4])
pd.DataFrame(data,columns=['Athlete','No of gold medals']).sort_values(by='No of gold medals',ascending=False).head().plot(x='Athlete',y='No of gold medals',kind='bar',title='Player with number of Gold medals',color='purple',figsize=(10,5))

  
```

# 6. In which year India won first Gold Medal in Summer Olympics?

```python
x=df[df['Country']=='IND']
year=x[x['Medal']=='Gold']['Year'].min()
print("India won first Gold Medal in Summer Olympics : ",year)
```

# 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]
for Event in df['Event'].unique():
  te=df[df['Event']==Event]
  c=0
  for Athlete in te['Athlete']:
    c+=1
  data.append([Event,c])

pd.DataFrame(data,columns=['Event','No of players played']).sort_values(by='No of players played',ascending=False).head().plot(x='Event',y='No of players played',kind='bar',color='gold',figsize=(10,5))
```

# 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data=[]
for Sport in df['Sport'].unique():
  te=df[df['Sport']==Sport]
  te1=te[te['Gender']=='Women']
  te2=te1[te1['Medal']=='Gold']
  data.append([Sport,len(te2)])
#print(data)
pd.DataFrame(data,columns=['Sport','No of female gold medalists']).sort_values(by='No of female gold medalists',ascending=False).head().plot(x='Sport',y='No of female gold medalists',kind='bar',color='red',title='sport with female gold medalists',figsize=(8,5))
```

# Cities with number of summer olympics conduction
